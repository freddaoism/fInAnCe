```python
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import yahoo_fin.stock_info as si
```

```python
F = wb.DataReader('F', data_source='yahoo', start='1995-1-1')
```

```python
F['log_return'] = np.log(F['Adj Close'] / F['Adj Close'].shift(1))
print (F['log_return'])
```

```python
F['log_return'].plot(figsize=(15,5))
plt.show()
```

```python
log_return_a = F['log_return'].mean() * 250
log_return_a
```

```python
#Apple Log return in percentage points
print (round(log_return_a, 5) * 100)
```
